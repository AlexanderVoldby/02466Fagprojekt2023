```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = "NMF"
data = "oil"

components = []

x = []
for i in range(1,14):
    x.append(np.min(np.load(data+"_"+model+"_"+str(i)+"_lr_test.npy"),axis=1))
    components.append(str(i))
x = np.array(x).T
#x = np.min(x,axis=2).T

lrs = ["1", "0.1", "0.01"]
x = x[1:]
lrs = lrs[1:]

plt.title("Model: "+str(model)+" Dataset: "+str(data))
plt.xlabel("nr of components")
plt.ylabel("loss")
for i,run in enumerate(x):
    plt.plot(components,run, label="lr:"+lrs[i])
plt.legend()
plt.show()
```
