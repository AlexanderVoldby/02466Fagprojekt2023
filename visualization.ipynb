 Trying to show difference between time and frequency domain. We want to correct a shift in the frequency domain(pseudo time domain), so we transform it into the pseudo frequency domain and make the shift there. 

```python
plt.style.use('ggplot')

#plt.plot(X[0])
#plt.show()
plt.title("Time domain of signal from NMR machine")
plt.plot(np.append(ifft(X[3])[100000:],ifft(X[3])[:20000]))
plt.show()
plt.title("Frequency domain of signal from NMR machine. Pseudo time domain")
plt.plot(X[3])
plt.show()
# plt.title("Pseudo frequency domain")
# plt.plot(np.append(fft(X[3])[100000:],fft(X[3])[:20000]))
# plt.show()
```

Try to introduce a shift in the pseudo time domain

```python
X = X[3][65000:85000]
```

```python
plt.title("Frequency domain of signal from NMR machine. Pseudo time domain")
plt.plot(X)
plt.show()
```

```python
plt.plot(X)
plt.show()
X_shift = np.array([fft(X)[f]*np.exp(-1j*2*pi*f*1000/len(X)) for f in tqdm(range(len(fft(X))))])

plt.title("")
plt.plot(ifft(X_shift))
plt.show()
```

```python
X_shift2 = shift(fft(X[3]),100, period = 1)

plt.title("")
plt.plot(ifft(X_shift2))
plt.show()
```
