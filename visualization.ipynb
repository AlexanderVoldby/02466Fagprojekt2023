```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
from helpers.data import X, labels
```

 Trying to show difference between time and frequency domain. We want to correct a shift in the frequency domain(pseudo time domain), so we transform it into the pseudo frequency domain and make the shift there. 

```python
plt.style.use('ggplot')

#plt.plot(X[0])
#plt.show()
plt.title("Time domain of signal from NMR machine")
plt.plot(np.append(ifft(X[3])[100000:],ifft(X[3])[:20000]))
plt.show()
plt.title("Frequeny domain of signal from NMR machine. Pseudo time domain")
plt.plot(X[3])
plt.show()
plt.title("Abstract/pseudo frequency domain. \"shift domain\" ")
plt.plot(np.append(fft(X[3])[100000:],fft(X[3])[:20000]))
plt.show()
```
